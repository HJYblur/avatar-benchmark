### Feature extraction

```python
from src.efficientnetFPN import EfficientNetV2FPN
from PIL import Image
from torchvision.transforms import Compose, Resize, ToTensor, Normalize


device = 'cuda' if torch.cuda.is_available() else 'cpu'

encoder = EfficientNetV2FPN(out_channels=256).to(device)

def load_image(path, device):
    img = Image.open(path).convert("RGB")
    img.show()
    transform = Compose([
        Resize((512, 512)),
        ToTensor(),
    ])
    img = transform(img).unsqueeze(0).to(device)  # Add batch dimension
    return img

data_path = "./data/dance.jpg"
img = load_image(data_path, device)

# Forward pass
with torch.no_grad():
    dense_map, deep_features = encoder(img)

print(f"Input: {img.shape}")
print(f"Dense Map: {dense_map.shape}") # Should be [1, 256, 128, 128]
print(f"Deep Features: {deep_features.shape}") # Should be [1, 1280, 16, 16]
```
